```python
import numpy as np
from numpy import asarray as npa
import pylab
from matplotlib import pyplot as plt
import matplotlib.ticker as mtick

%matplotlib inline
```

```python
#Average Case
T = npa([[0,0],[5000,5001],[10000,9000],[15000,16001],[20000,19001],[25000,25002],[30000,29002]])
```

```python
fig = plt.figure()
plt.ylim((0,4e4))
plt.ylabel("Time in us")
plt.xlabel("Nuber Of Elements")
ax = fig.add_subplot(111)
ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.e'))
plt.tight_layout()
plt.plot(T[:,0],T[:,1])
pylab.savefig('average.png')
```
