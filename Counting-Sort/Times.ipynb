```python
import numpy as np
from numpy import asarray as npa
import pylab
from matplotlib import pyplot as plt
import matplotlib.ticker as mtick

%matplotlib inline
```

```python
#Average Case
T = npa([[0,0], [500000,15600], [1000000,62400], [1500000,78001], [2000000,109201], [2500000,124800], [3000000,156001]])
```

```python
fig = plt.figure()
plt.ylim((0,2e5))
plt.ylabel("Time in us")
plt.xlabel("Nuber Of Elements")
ax = fig.add_subplot(111)
ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.e'))
plt.tight_layout()
plt.plot(T[:,0],T[:,1])
pylab.savefig('average.png')
```
