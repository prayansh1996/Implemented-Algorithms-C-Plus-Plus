```python
import numpy as np
from numpy import asarray as npa
import pylab
from matplotlib import pyplot as plt
from scipy.interpolate import spline
import matplotlib.ticker as mtick

%matplotlib inline
```

```python
#Average Case
T = npa([[0,0], [5000,230000], [10000,921001], [15000,2060003], [20000,3665034], [25000,5698024]])
```

```python
fig = plt.figure()
plt.ylim((0,7e6))
plt.ylabel("Time in seconds")
plt.xlabel("Nuber Of Elements")
ax = fig.add_subplot(111)
ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.e'))
plt.tight_layout()
plt.plot(T[:,0],T[:,1])
#pylab.savefig('average.png')
```
