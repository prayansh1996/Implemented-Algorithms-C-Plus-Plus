```python
import numpy as np
from numpy import asarray as npa
import pylab
from matplotlib import pyplot as plt
import matplotlib.ticker as mtick

%matplotlib inline
```

```python
#Average Case
T = npa([[0,0], [5000,109200], [10000,431813], [15000,790606], [20000,1809603], [25000,2411418], [30000,4425210]])
```

```python
fig = plt.figure()
plt.ylim((0,6e6))
plt.ylabel("Time in seconds")
plt.xlabel("Nuber Of Elements")
ax = fig.add_subplot(111)
ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.e'))
plt.tight_layout()
plt.plot(T[:,0],T[:,1])
#pylab.savefig('average.png')
```
